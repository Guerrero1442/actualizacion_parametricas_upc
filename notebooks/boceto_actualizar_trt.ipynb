```python
import pathlib
from unidecode import unidecode
import pandas as pd
from tkinter import filedialog
```

```python
ruta_trt = pathlib.Path(filedialog.askopenfilename())
ruta_trt
```

```python
df = pd.read_excel(ruta_trt, dtype=str)
df.head()
```

```python
def quitar_acentos(texto: str) -> str:
    return unidecode(texto)
```

```python
# obtener solo el nombre de las columnas del archivo
df = pd.read_excel(ruta_trt, nrows=0)

# Limpiar nombres de columnas
# quitar acentos de las columnas
df.columns = [quitar_acentos(col).strip().upper() for col in df.columns]

    
# remplazar espacios dobles de las columnnas
df.columns = df.columns.str.replace(r'\s{2,}', ' ', regex=True).str.strip()
        
        
# Agregar guion bajo antes de paréntesis de apertura si no está precedido por uno
df.columns = df.columns.str.replace(r'([^\_])\(', r'\1_(', regex=True)

# Agregar guion bajo antes de paréntesis de cierre si no está precedido por uno Y está seguido de un carácter
df.columns = df.columns.str.replace(r'([^\_])\)(?=\w)', r'\1_)', regex=True)

# Luego eliminar los paréntesis como en tu código original si lo deseas
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')

        
# remplazar espacios por guiones bajos
df.columns = df.columns.str.replace(' ', '_')

columnas = df.columns.to_list()
```

```python
type(columnas)
```

```python
df = pd.read_excel(ruta_trt, dtype='str', names=columnas)

df.columns = df.columns.str.replace('.', '')

df.columns.to_list()
```

```python
df.head()
```

```python
# df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
for col in df.columns:
    try:
        df[col] = df[col].str.strip()
    except AttributeError:
        print(f'Error en columna {col}')
```

```python
columnas_longitud_maxima = {col: df[col].astype(str).map(len).max() for col in df.columns}
columnas_longitud_maxima
```

```python
nombre_tabla = input('Nombre de la tabla: ').lower()
nombre_tabla
```

```python
def crear_tabla(nombre_tabla: str, dataframe: pd.DataFrame, columnas: list = None, longitud_columnas: dict = None) -> str:

    if columnas is None:
        columnas = dataframe.columns.tolist()
    if longitud_columnas is None:
        longitud_columnas = {}

    columnas_definicion = []
    for columna in columnas:
        cantidad_bytes_columna = longitud_columnas.get(columna, 255)
        columnas_definicion.append(f"{columna} VARCHAR2({cantidad_bytes_columna})")

    columnas_definicion_str = ",".join(columnas_definicion)
    sentencia_sql = f"CREATE TABLE {nombre_tabla} ({columnas_definicion_str})"
    return sentencia_sql



sentencia_sql = crear_tabla(nombre_tabla, df, longitud_columnas=columnas_longitud_maxima)

sentencia_sql
```

```python
import cx_Oracle
import configparser

# Leer el archivo de configuración
CONFIG_PATH = '../config.ini'
config = configparser.ConfigParser()
config.read(CONFIG_PATH)

# Extraer los datos de conexión
host = config['oracle']['host']
port = config['oracle']['port']
sid = config['oracle']['service_name']
user = config['oracle']['user']
password = config['oracle']['password']

# Crear el DSN para la conexión
dsn = cx_Oracle.makedsn(host, port, sid)


# Imprimir la sentencia SQL para verificarla
print("Sentencia SQL que se va a ejecutar:")
print(sentencia_sql)

```

```python
# Establecer la conexión y ejecutar la sentencia
try:
    connection = cx_Oracle.connect(user, password, dsn)
    cursor = connection.cursor()

    # Ejecutar la sentencia SQL
    cursor.execute(sentencia_sql)
    connection.commit()  # Confirmar los cambios
    print("Tabla creada exitosamente")

except cx_Oracle.DatabaseError as e:
    error, = e.args
    print(f'Error al ejecutar sentencia SQL: {error.message}')
finally:
    # Cerrar el cursor y la conexión
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```

```python
from sqlalchemy import create_engine


try:
    connection_string = f'oracle+cx_oracle://{user}:{password}@{host}:{port}/{sid}'
    engine = create_engine(connection_string)
    connection = engine.connect()
    print("Conexión exitosa")
    # Insertar los datos del DataFrame en la tabla
    df.to_sql(nombre_tabla, con=engine, if_exists='append', index=False)
except:
    print("Error al conectar a la base de datos")
finally:
    if connection:
        connection.close()
    if engine:
        engine.dispose()


print("Datos insertados exitosamente")
```
